```python
#Data Manipulation
import pandas as pd
data={
    "Name":["A","B","C"],
    "Age":[22,25,23],
    "Marks":[85,78,92]
}
df=pd.DataFrame(data)
print("Original Data:")
print(df)
print("\nAdding a new column 'Bonus'\n")
df["Bonus"]=5
print(df)
print("\nModifying a column 'Marks'\n")
df["Marks"]=df["Marks"]+5
print(df)
print("\nCreating a new column based on other columns\n")
df["Total Marks"]=df["Marks"]+df["Bonus"]
print(df)
print("\nRenaming a column\n")
df.rename(columns={"Total":"Final"},inplace=True)
print(df)
print("\nDropping a column\n")
df.drop("Bonus",axis=1,inplace=True)
print(df)
print("\nChanging a particular value in the dataframe\n")
df.loc[0,"Age"]=21
print(df)
```
