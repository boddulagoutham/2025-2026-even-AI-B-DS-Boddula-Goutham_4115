```python
import numpy as np
import math
data = np.array([10, 15, 20, 25, 30])
n = len(data)
print("data size (n):", n)
mean_data = np.mean(data)
print("mean (μ):", mean_data)
sigma_data = np.std(data, ddof=0)
print("standard deviation (σ):", sigma_data)
z_scores = (data - mean_data)/sigma_data
print("standardized values (z-scores):", z_scores)
```
