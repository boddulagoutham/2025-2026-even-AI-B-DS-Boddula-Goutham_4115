```python
import numpy as np
import math
x = np.array([0,1,2,3,4])
p = np.array([0.1,0.2,0.3,0.25,0.15])
n = len(x)
print("number of outcomes:",n)
print("values of X:",x)
print("probabilities:",p)
sum_p = np.sum(p)
print("sum of probabilities:",sum_p)
mu = np.sum(x*p)
print("mean (E[X]):",mu)
sigma2 = np.sum((x-mu)**2 * p)
print("variance:",sigma2)
sigma = math.sqrt(sigma2)
print("standard deviation:",sigma)
```
