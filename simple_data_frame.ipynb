```python
# Get number of elements
n = int(input("Enter number of elements: "))

# Initialize empty list
arr = []

# Get elements from the user
for i in range(n):
    element = int(input(f"Enter element {i+1}: "))
    arr.append(element)

# Get the element to search for
key = int(input("Enter element to search: "))

# Initialize found flag
found = False

# Search for the element
for i in range(n):
    if arr[i] == key:
        print(f"Element {key} found at position {i+1}")
        found = True
        break

# If element was not found
if not found:
    print("Element not found")
```

```python
import pandas as pd

# Simple DataFrame
data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [80, 85, 75, 90, 88]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

print("\nAverage Marks by department:")
print(df.groupby("Department")["Marks"].mean())

print("\nTotal marks by department:")
print(df.groupby("Department")["Marks"].sum())

print("\nStudent count by department:")
print(df.groupby("Department")["Marks"].count())
```
