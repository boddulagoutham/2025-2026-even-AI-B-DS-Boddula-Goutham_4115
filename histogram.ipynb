```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(1, 20, 50)

plt.figure(figsize=(6,4))
plt.hist(data, bins=8, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

```
