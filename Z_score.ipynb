```python

import numpy as np
data = [10,20,30,40,50]
mean = np.mean(data)
print("mean:",mean)
std_dev=np.std(data,ddof=1)
print("standard Deviation:",std_dev)
x = 40
Z_score=(x-mean)/std_dev
print("Z-score of 40:",Z_score)
Z_scores = [(x-mean)/ std_dev for x in data]
print("Z-scores",Z_scores)
```
