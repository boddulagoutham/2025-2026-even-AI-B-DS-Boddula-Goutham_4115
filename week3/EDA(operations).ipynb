```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
values = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]
df = pd.DataFrame(values, columns=["values"])

# Display dataset
print("Dataset:")
print(df)

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Variance
variance = np.var(values, ddof=1)  # sample variance
print("\nVariance:", round(variance, 2))

# Standard deviation
std_dev = np.std(values, ddof=1)  # sample standard deviation
print("Standard deviation:", std_dev)

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(values, bins=5, edgecolor='black')
plt.title("Data Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
